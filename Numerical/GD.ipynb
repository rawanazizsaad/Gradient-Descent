```python
import numpy as np
```

```python

# Step 1: Initialize parameters and learning rate
theta_0 = 0.0
theta_1 = 0.0
learning_rate = 0.01
max_iterations = 1000
convergence_threshold = 1e-5

```

```python
# Step 2: Prediction function
def predict(x, theta_0, theta_1):
    return theta_0 + theta_1 * x

```

```python
# Step 3: Cost function
def calculate_cost(x, y, theta_0, theta_1):
    m = len(x)
    h_x = predict(x, theta_0, theta_1)
    cost = (1 / (2 * m)) * np.sum((h_x - y) ** 2)
    return cost

```

```python
# Step 4: Gradient calculation
def calculate_gradient(x, y, theta_0, theta_1):
    m = len(x)
    h_x = predict(x, theta_0, theta_1)
    d_theta_0 = (1 / m) * np.sum(h_x - y)
    d_theta_1 = (1 / m) * np.sum((h_x - y) * x)
    return d_theta_0, d_theta_1

```

```python
# Step 5: Update parameters
def update_parameters(theta_0, theta_1, d_theta_0, d_theta_1, learning_rate):
    theta_0 -= learning_rate * d_theta_0
    theta_1 -= learning_rate * d_theta_1
    return theta_0, theta_1

```

```python
# Step 6: Gradient Descent loop
def gradient_descent(x, y, theta_0, theta_1, learning_rate, max_iterations, convergence_threshold):
    iteration = 0
    while iteration < max_iterations:
        cost = calculate_cost(x, y, theta_0, theta_1)
        d_theta_0, d_theta_1 = calculate_gradient(x, y, theta_0, theta_1)
        
        if abs(cost) < convergence_threshold:
            print(f"Converged to minimum at iteration {iteration}, Cost: {cost}")
            break
        
        theta_0, theta_1 = update_parameters(theta_0, theta_1, d_theta_0, d_theta_1, learning_rate)
        
        if iteration % 100 == 0:
            print(f"Iteration {iteration}, Cost: {cost}")
        
        iteration += 1
    
    return theta_0, theta_1

```

```python
# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Running the gradient descent algorithm
theta_0, theta_1 = gradient_descent(x, y, theta_0, theta_1, learning_rate, max_iterations, convergence_threshold)

# Print the final learned parameters
print("Learned Parameters:")
print("theta_0:", theta_0)
print("theta_1:", theta_1)
```
